```python
import pandas as pd
import numpy as np
```

```python
# grab data
dbscan_v1 = pd.read_csv('../../data/evaluation_data/dbscan_clusters.csv', index_col=0).drop('borough', axis=1)
dbscan_v2 = pd.read_csv('../../data/evaluation_data/dbscan_v2_clusters.csv', index_col=0).drop('borough', axis=1)
dbscan_v3 = pd.read_csv('../../data/evaluation_data/dbscan_v3_clusters.csv', index_col=0).drop('borough', axis=1)

costs = pd.read_csv('../../data/evaluation_data/cluster_housing_costs.csv', index_col=0)
```

```python
costs
```

```python
# add economic data
dbscan_v1 = pd.concat([dbscan_v1, costs['average_costs']], axis=1)
dbscan_v2 = pd.concat([dbscan_v2, costs['average_costs']], axis=1)
dbscan_v3 = pd.concat([dbscan_v3, costs['average_costs']], axis=1)
```

```python
print(dbscan_v1.groupby('cluster').count())
dbscan_v1.groupby('cluster').mean()
```

```python
print(dbscan_v2.groupby('cluster').count())
dbscan_v2.groupby('cluster').mean()
```
