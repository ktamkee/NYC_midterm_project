```python
import pandas as pd
from pandas.io.json import json_normalize
import requests as re
import os
```

```python
foursquare_secret = os.environ['FOURSQUARE_API_KEY']
API_KEY = os.environ["YELP_API_KEY"]
```

```python
main_dataset = pd.read_csv('main_dataset.csv')
```

```python
bronx_neighborhoods = main_dataset[['borough', 'neighborhood']][main_dataset['borough']=='Bronx']
bronx_neighborhoods_lst = bronx_neighborhoods.neighborhood.tolist()
len(bronx_neighborhoods)
```

```python
bronx = pd.DataFrame()

for i in bronx_neighborhoods_lst:
    string = f'{i}, NY'
    url = "http://api.foursquare.com/v3/places/search"
    headers = {'Accept': 'application/json'}
    headers['Authorization']= foursquare_secret
    rest_parameters = {
        'near': string,
        'sort': 'POPULARITY',
        'limit': '50'
    }

    response = re.get(url, params=rest_parameters, headers = headers)
    response = response.json()
    venues = pd.json_normalize(response, record_path=['results'])

    bronx = pd.concat([bronx, venues])
```

```python
bronx = bronx[['location.neighborhood', 'location.locality', 'categories']]
bronx = bronx.dropna()
bronx = bronx[bronx["categories"].str.len() != 0]
bronx.head()
```

```python
bronx_names = bronx['location.neighborhood'].tolist()
bronx_cat = bronx['categories'].tolist()
```

```python
bronx_names2 = []
for i in bronx_names:
   bronx_names2.append(i[0].strip())

bronx_cat2 = []
for i in bronx_cat:
    bronx_cat2.append(i[0]['name'])
```

```python
bronx_df = pd.DataFrame()
bronx_df['neighborhood'] = bronx_names2
bronx_df['categories'] = bronx_cat2
```

```python
bronx_df['neighborhood'] = bronx_df['neighborhood'].str.replace('Coop City', 'Co-op City')
bronx_df['neighborhood'] = bronx_df['neighborhood'].str.replace('Co-Op City', 'Co-op City')
bronx_df.head()
```
