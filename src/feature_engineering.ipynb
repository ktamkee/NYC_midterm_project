```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
main_data = pd.read_csv('../data/main_dataset.csv')
```

```python
main_features = main_data.loc[:,'population':].drop(['restaurant_count','N'], axis=1)
```

```python
population = main_features['population']
population
```

```python
yelp_types = pd.read_csv('../data/data/yelp_restaurant_type_counts_1km.csv').drop('Unnamed: 0', axis=1)
yelp_types
```

```python
google_types = pd.read_csv('../data/data/all_venues_data_1km.csv').drop('id', axis=1)
google_types
```

```python
all_features = pd.concat([population, google_types, yelp_types], axis=1)
```

```python
all_features.to_csv('../data/data/all_features_counts.csv')
```

```python
all_features.head()
```

```python
# divide all rows by population
for row in range(all_features.shape[0]):
   all_features.iloc[row] = all_features.iloc[row].div(all_features.iloc[row]['population'])
    
```

```python
all_features
```

```python
# remove population
all_features = all_features.drop('population', axis=1)
```

```python
all_features
```

```python
# scaling
scaler = MinMaxScaler()
all_features_scaled = pd.DataFrame(scaler.fit_transform(all_features), columns=all_features.columns)
all_features_scaled
```

```python
# export
all_features_scaled.to_csv('../data/data/all_features_scaled.csv', index=False)
```
