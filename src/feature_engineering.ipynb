```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
main_data = pd.read_csv('../data/main_dataset.csv')
```

```python
main_features = main_data.loc[:,'population':].drop(['restaurant_count','N'], axis=1)
```

```python
population = main_features['population']
population
```

```python
yelp_types = pd.read_csv('../data/data/yelp_restaurant_type_counts_1km.csv').drop('Unnamed: 0', axis=1)
yelp_types
```

```python
google_types = pd.read_csv('../data/data/google_counts_v3.csv', index_col=0)
google_types
```

```python
all_features = pd.concat([population, google_types, yelp_types], axis=1)
```

```python
all_features.to_csv('../data/data/all_features_counts_v3.csv')
```

```python
main_features = all_features.drop('index', axis=1)
```

```python
# divide all rows by population
for row in range(main_features.shape[0]):
   main_features.iloc[row] = main_features.iloc[row].div(main_features.iloc[row]['population'])
    
```

```python
main_features
```

```python
# remove population
main_features = main_features.drop('population', axis=1)
```

```python
main_features
```

```python
# scaling
scaler = StandardScaler()
main_features_scaled = pd.DataFrame(scaler.fit_transform(main_features), columns=main_features.columns)
main_features_scaled
```

```python
# export
main_features_scaled.to_csv('../data/data/all_features_scaled.csv', index=False)
```
