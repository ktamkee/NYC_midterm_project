```python
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
file = open('data/google_restaurants.json')
data = json.load(file)
data = pd.DataFrame(data).T
data
```

```python
data = data.reset_index(level=0)
data.columns=['id', 'restaurant_counts', 'restaurant_review_counts', 'restaurant_ratings', 'restaurant_price_levels']
data.head()
```

```python
restuarants_data = pd.DataFrame(data['id'])
restuarants_data['number_of_restaurants'] = data['restaurant_counts'].apply(sum)
```

```python
review_count_lst = data.restaurant_review_counts.tolist()
average_review_count = []

for lst in review_count_lst:
    if len(lst) > 0:
        average_review_count.append(round(sum(lst) / len(lst)))
    else:
        average_review_count.append(0)

restuarants_data['average_review_count'] = average_review_count
```

```python
restautant_ratings_lst = data.restaurant_ratings.tolist()
average_rating_lst = []

for lst in restautant_ratings_lst:
    if len(lst) > 0:
        average_rating_lst.append(round(sum(lst) / len(lst),1))
    else:
        average_rating_lst.append(0)

restuarants_data['average_restaurant_rating'] = average_rating_lst
```

```python
price_level_lst = data.restaurant_price_levels.tolist()
average_price_level = []

for lst in price_level_lst:
    numerical_lst = []
    for i in lst:
        if type(i) == int:
            numerical_lst.append(i)
    if len(numerical_lst) > 0:
        average_price_level.append(round(sum(numerical_lst) / len(numerical_lst)))
    else:
        average_price_level.append(0)

restuarants_data['average_price_level'] = average_price_level
```

```python
venue_file = open('data/google_all_venues.json')
venue_data = json.load(venue_file)
venue_data = pd.DataFrame(venue_data).T
venue_data
```

```python
venue_data = venue_data.reset_index(level=0)
```

```python
venue_data.columns=['id', 'counts', 'review_counts', 'ratings', 'type', 'price']
venue_data['venue_count'] = venue_data['counts'].apply(sum)
venue_data.head()
```

```python
venue_category_list = list()

for venue in venue_data.type:
    
    temp_list = []

    for i in venue:
        temp_list.append(i[0])

    venue_category_list = venue_category_list + temp_list
```

```python
venue_counts = venue_data.venue_count.to_list()
id = venue_data.id.to_list()
```

```python
neighborhood_id = []
for (id, count) in zip(id, venue_counts):
     for i in range(count):
         neighborhood_id.append(id)
```

```python
print(len(neighborhood_id))
print(len(venue_category_list))
```

```python
venues = pd.DataFrame()
venues['id'] = neighborhood_id
venues['venue_category'] = venue_category_list
venues.head()
```

```python
onehot = pd.get_dummies(venues[['venue_category']], prefix='', prefix_sep='')
venues = venues.join(onehot)
venues = venues.drop(columns=['venue_category'], axis = 1)
venues.head()
```

```python
venues = venues.groupby('id').sum()
venues.head()
```
