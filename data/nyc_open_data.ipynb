```python
# Load nyc_geo JSON
import json
import pandas as pd
from pandas.io.json import json_normalize

file = open('nyc_geo.json')
data = json.load(file)
```

```python
df = pd.json_normalize(data['features'])

# Tidy Data
nyc_df = df[['properties.name', 'properties.borough']]
nyc_df.columns = ['neighborhood', 'borough']
nyc_df
```

```python
nyc_df.drop_duplicates(keep=False)
```

```python
nyc_lst = nyc_df['neighborhood'].to_list()
```

```python
population_lst = population['NTA Name'].to_list()
```

```python
main_list = [neighborhood for neighborhood in population_lst if neighborhood not in nyc_lst]
```
